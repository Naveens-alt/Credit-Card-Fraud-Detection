```python
#importing

import numpy as np
import pandas as pd 
from sklearn.model_selection import train_test_split  #allow data to train and test
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#read csv file

credit_card_file = pd.read_csv('C:/Users/naveen.s8/Documents/Project/creditcard fraud/creditcard.csv/creditcard.csv')
```

```python
#head (show first 5 files)
credit_card_file.head()
```

```python
#tail (show last 5 files)

credit_card_file.tail()
```

```python
#show imnformation

credit_card_file.info()
```

```python
#show non null values

credit_card_file.isnull().sum()
```

```python
#UNBALANCED

#distribution of legit transaction and fraud transaction

credit_card_file['Class'].value_counts()
```

```python
#Seperating the data for analysis

legit = credit_card_file[credit_card_file.Class == 0]
fraud = credit_card_file[credit_card_file.Class == 1]
```

```python
#print the seperating data

print(legit.shape)
print(fraud.shape)
```

```python
#Statstical measure

legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the value of both transaction

credit_card_file.groupby('Class').mean()
```

```python
#under sampling
legit_sample = legit.sample(492)
```

```python
#concat

new_data = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_data['Class'].value_counts()
```

```python
new_data.describe()
```

```python
x = new_data.drop(columns= 'Class', axis=1)
y = new_data['Class']
```

```python
#split the data for training and test

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, stratify=y, random_state= 2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
#MODEL RAINING ACCURACY

#Logistic Regression

model = LogisticRegression()
```

```python
#training the LRM with training data

model.fit(x_train, y_train)
```

```python
#EVALUATION

x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print(training_data_accuracy)
```

```python
#accuracy on test data

x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print(test_data_accuracy)
```
